```python
variaveis = ['SEXO', 'QUANTAS PESSOAS MORAM COM VOCÊ', 'RENDA/FAMILIA', 'ESCOLARIDADE/PAIS', 'ESTADO CIVIL',
             'QUANTOS FILHOS', 'TRABALHA', 'RAÇA', 'TIPO DE ESCOLA', 'CURSO DE INTERESSE', 'TURNO INTERESSE',
             'GASTO COM INTERNET DIÁRIO', 'VOCÊ ESTUDARIA NESSA FACULDADE']
```

#VISUALIZAÇÃO

---



```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
variaveis = [variavel for variavel in variaveis if variavel != ['IDADE' , 'CURSO DE INTERESSE']]


for variavel in variaveis:
    print(f'\nVariável: {variavel}')
    print(f'Categorias Únicas: {df[variavel].unique()}')
    print(f'Tabela de Frequência:\n{df[variavel].value_counts(dropna=False)}\n')

    if df[variavel].nunique() <= 10:
        plt.figure(figsize=(6, 4))
        sns.countplot(y=variavel, data=df, palette='viridis')
        plt.title(f'Distribuição de {variavel}')
        plt.show()
```

###ANÁLISE DOS CURSOS DE PREFERÊNCIA

---



```python
top_cursos = df['CURSO DE INTERESSE'].value_counts().head(20)


plt.figure(figsize=(10, 6))
sns.barplot(x=top_cursos.values, y=top_cursos.index, palette='viridis')
plt.title('Top 20 Cursos de Interesse')
plt.xlabel('Número de Respostas')
plt.ylabel('Curso')
plt.show()
```

```python
top_cursos_masculinos = df[df['SEXO'] == 'Masculino']['CURSO DE INTERESSE'].value_counts().head(20)
top_cursos_femininos = df[df['SEXO'] == 'Feminino']['CURSO DE INTERESSE'].value_counts().head(20)


plt.figure(figsize=(10, 6))
sns.barplot(x=top_cursos_masculinos.values, y=top_cursos_masculinos.index, palette='viridis')
plt.title('Top 20 Cursos de Interesse para Homens')
plt.xlabel('Número de Respostas')
plt.ylabel('Curso')
plt.show()


plt.figure(figsize=(10, 6))
sns.barplot(x=top_cursos_femininos.values, y=top_cursos_femininos.index, palette='viridis')
plt.title('Top 20 Cursos de Interesse para Mulheres')
plt.xlabel('Número de Respostas')
plt.ylabel('Curso')
plt.show()
```

## ANALISE BIVARIADA

---
SEXO X ESTADO CIVIL


```python
tab = pd.crosstab(df['SEXO'], df['ESTADO CIVIL'], margins= True)
tab
```

```python
df_tab = df.groupby('SEXO')['ESTADO CIVIL'].value_counts().unstack()
df_tab.plot(kind='barh', stacked=True)
plt.show()
```

SEXO X TRABALHA

```python
tab = pd.crosstab(df['SEXO'], df['TRABALHA'], margins= True)
tab
```

```python
df_tab = df.groupby('SEXO')['TRABALHA'].value_counts().unstack()

df_tab.plot(kind='bar', stacked=True  )

plt.show()
```
